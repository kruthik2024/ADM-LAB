```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

area = np.array([1000, 1500, 2000]).reshape(-1, 1)
price = np.array([3000000, 4500000, 6000000])

model = LinearRegression()
model.fit(area, price)

slope = model.coef_[0]
intercept = model.intercept_
prediction = model.predict([[1700]])

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Predicted price for 1700 sq ft: {prediction[0]} Rs")

mae = mean_absolute_error(price, model.predict(area))
r2 = r2_score(price, model.predict(area))

print(f"Mean Absolute Error: {mae}")
print(f"R² Score: {r2}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np

study_hours = np.array([2, 4, 6]).reshape(-1, 1)
attendance = np.array([60, 80, 90])
passed = np.array([0, 1, 1])

X = np.column_stack((study_hours, attendance))
y = passed

model = LogisticRegression()
model.fit(X, y)

prediction = model.predict([[5, 85]])
print(f"Prediction for 5 study hours and 85% attendance: {'Pass' if prediction[0] == 1 else 'Fail'}")

accuracy = accuracy_score(y, model.predict(X))
print(f"Model Accuracy: {accuracy}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

y_true = np.array([1, 0, 1, 1, 0, 1])
y_pred = np.array([1, 0, 0, 1, 0, 1])

precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
accuracy = accuracy_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

age = np.array([22, 35, 28]).reshape(-1, 1)
salary = np.array([30000, 60000, 50000])
purchased = np.array([0, 1, 1])

X = np.column_stack((age, salary))
y = purchased

model = DecisionTreeClassifier()
model.fit(X, y)

plt.figure(figsize=(10, 6))
plot_tree(model, filled=True, feature_names=["Age", "Salary"], class_names=["No", "Yes"], rounded=True)
plt.show()

prediction = model.predict([[30, 55000]])
print(f"Prediction for 30-year-old with 55,000 salary: {'Purchase' if prediction[0] == 1 else 'No Purchase'}")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

income = np.array([25, 27, 50, 54, 23, 52]).reshape(-1, 1)
spending_score = np.array([30, 32, 70, 72, 29, 69])

X = np.column_stack((income, spending_score))

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('K-Means Clustering (k=2)')
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

transactions = [
    ['milk', 'bread', 'butter'],
    ['bread', 'diaper', 'beer'],
    ['milk', 'bread', 'diaper', 'beer'],
    ['bread', 'milk', 'diaper', 'butter']
]

df = pd.DataFrame(transactions, columns=['item1', 'item2', 'item3', 'item4']).fillna('')

df_bool = df.applymap(lambda x: x != '').astype(int)

frequent_itemsets = apriori(df_bool, min_support=0.5, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
print("Association Rules:")
print(rules)

rule = rules.iloc[0]
print(f"Rule: {rule['antecedents']} -> {rule['consequents']}")
```
